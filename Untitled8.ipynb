```python
from fuzzywuzzy import fuzz
import sklearn
from sklearn.metrics import precision_recall_fscore_support
y_test=[]
y_pred=[]

output=np.zeros((X_test,6))
tp, fp, fn = 0, 0, 0
with open(file_path4, 'r') as f2, open(file_path3, 'r') as f1:
  for index in range(X_test):
    a_=f2.readline()
    b_=f1.readline()
    b=b_.split(',')
    a=a_.split(' ')
    for i in range(6):
      a[i]=a[i].lower()
      b[1]=b[1].lower()
      a[i]=a[i].replace('_', ' ')
      b[1]=b[1].replace('_', ' ')
      a[i]=a[i].strip()
      b[1]=b[1].strip()
      output[index,i]=fuzz.partial_ratio(a[i], b[1])
    y_test.append(a[0])
    y_pred.append(b[1])
    #print(a[0], b[1],output[index,0])
    if output[index,0]>=80 or output[index,1]>=80 or output[index,2]>=80:
      tp+=1
    else:
      fp+=1

fn = X_test - tp

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * ((precision * recall) / (precision + recall))

#precision = precision_recall_fscore_support(y_test, y_pred,average='macro')
#recall = recall_score(y_test, y_pred,average='micro')
#f1_score = f1_score(y_test, y_pred,average='micro')

print('hit1' ,np.count_nonzero(output[:,0] >= 80)/X_test)
print('hit3',(np.count_nonzero(output[:,0] >= 80)+np.count_nonzero(output[:,1] >= 80)+np.count_nonzero(output[:,2] >= 80))/X_test)
print('hit5' ,np.count_nonzero(output[:,4] >= 80)/X_test)
MR=np.count_nonzero(output[:,0] >= 80)+2*np.count_nonzero(output[:,1] >= 80) +3*np.count_nonzero(output[:,2] >= 80)+4*np.count_nonzero(output[:,3] >= 80)+5*np.count_nonzero(output[:,4] >= 80) +6*np.count_nonzero(output[:,5] >= 80)
MRR=1/(np.count_nonzero(output[:,0] >= 80))+1/(2*np.count_nonzero(output[:,1] >= 80)) +1/(3*np.count_nonzero(output[:,2] >= 80))+1/(4*np.count_nonzero(output[:,3] >= 80))+1/(5*np.count_nonzero(output[:,4] >= 80)) +1/(6*np.count_nonzero(output[:,5] >= 80))

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1_score)
print('MR' ,MR/X_test)
print('MRR' ,MRR/X_test)

```
